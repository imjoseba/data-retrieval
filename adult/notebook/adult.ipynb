```python
import os
import re
import pandas as pd
import wget
```

```python
# Download the 'adult.data' file from a given URL.
wget.download('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
```

```python
# Download the 'adult.names' file from a given URL.
wget.download('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names')
```

```python
# Open the 'adult.data' file for reading.
with open(os.path.join(os.getcwd(),'adult.data'),'r') as f:
    # Read the data from the file and split it into lines.
    data = f.read().splitlines()
    # Split each line into a list using ', ' as the separator, effectively creating a list of lists.
    data = [elem.split(', ') for elem in data]
```

```python
# Open the 'adult.names' file for reading
with open(os.path.join(os.getcwd(),'adult.names'),'r') as f:
    # Read the metadata from the file and split it into lines.
    metadata = f.read().splitlines()
```

```python
# Define a lambda function to extract specific patterns from metadata lines.
regex_fn = lambda text: re.findall('^[a-zA-Z-]+:{1}', text)
# Define another lambda function to extract column names from metadata lines.
reg_text_fn = lambda text : re.findall('[a-zA-Z- ]+', text)
# Process the metadata and extract relevant information using the 'regex_fn' and 'reg_text_fn' functions.
metadata_list = [regex_fn(elem)[0] for elem in metadata if regex_fn(elem)]
# Extract the column names from the processed metadata and add 'label' as the last column name.
col_names = [reg_text_fn(elem)[0] for elem in metadata_list if reg_text_fn(elem)] + ['label']
```

```python
# Create a DataFrame from the processed data and column names.
df = pd.DataFrame(data=data, columns=col_names)
df
```
